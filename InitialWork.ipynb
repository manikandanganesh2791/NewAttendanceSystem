```python
attendance_s = open('attendance_s.csv', 'r')
row_s = attendance_s.read().splitlines()
row_list = []
record_dict = {}
date_split_list = []
date_val = ""
def list_val(row_list):
    result = []
    for i in range(len(row_list)):
        date_val = ""
        date_split_list = []
        if(i not in (0,4,5)):
            if(i==3):
                if(row_list[i] != 'NULL'):
                    date_split_list=row_list[i].split("/")
                    for j in range(len(date_split_list)):
                        if(j != 2):
                            date_val+=date_split_list[j]+'/'
                        else:
                            date_val+=date_split_list[j][0:4]
                    result.append(date_val)
                else:
                    result.append(row_list[i])
            elif(i==6):
                if(row_list[i] != 'NULL'):
                    date_split_list=row_list[i].split("/")
                    for j in range(len(date_split_list)):
                        if(j != 2):
                            date_val+=date_split_list[j]+'/'
                        else:
                            date_val+=date_split_list[j][0:4]
                    result.append(date_val)
                else:
                    result.append(row_list[i])
            else:
                result.append(row_list[i])
    return result
for line in row_s[1:]:
    row_list = line.split(",")
    if(row_list[1] == '73'):
        if(row_list[0] in record_dict):
            row_list =[]
            continue
        record_dict[row_list[0]] = list_val(row_list)
        row_list = []

list_dates = open('List of school days .csv', 'r')
row_dates = list_dates.read().splitlines()
dates_list = []

for row_dates_line in row_dates[1:]:
    dates_list.append(row_dates_line)
missing_dates = []
dates_set = set()

for dict_date in record_dict.values():
    if(dict_date[2] != 'NULL' and dict_date[3] != 'NULL'):
        dates_set.add(dict_date[2])
        dates_set.add(dict_date[3])
    elif(dict_date[2] != 'NULL' and dict_date[3] == 'NULL'):
        dates_set.add(dict_date[2])
    elif(dict_date[3] != 'NULL' and dict_date[2] == 'NULL'):
          dates_set.add(dict_date[3])

attendance_c = open('attendance_c.csv', 'r')
row_c = attendance_c.read().splitlines()

for line in row_c[1:]:
    row_list =[]
    row_list = line.split(",")
    dates_set.add(row_list[1])

attendance_k = open('attendance_k.csv', 'r')
row_k = attendance_k.read().splitlines()
for line in row_k[1:]:
    row_list =[]
    row_list = line.split(",")
    dates_set.add(row_list[3])

for dates_list_val in range(len(dates_list)):
    if(dates_list[dates_list_val] not in dates_set):
        missing_dates.append(dates_list[dates_list_val])

missing_dates
```
